# Exemplo básico de Regressão Linear

Contribuição de Filipe Dias apoiado pelo ChatGPT.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Carregamento de Dados

```python
# Carregar o conjunto de dados
df = pd.read_csv("dados_regressao.csv")

# Separar variáveis independentes (X) e dependentes (Y)
X = df[['X']]
Y = df['Y']
```

# Treinamento e Avaliação

```python

# Criar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X.values, Y)

# Obter coeficientes
a = modelo.coef_[0]
b = modelo.intercept_

print(f"Equação da reta: Y = {a:.2f}X + {b:.2f}")

# Fazer previsões
y_pred = modelo.predict(X.values)

# Avaliação do modelo
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)

print(f"Erro quadrático médio: {mse:.2f}")
print(f"Coeficiente de determinação (R²): {r2:.2f}")
```

# Predição

```python
previsto = modelo.predict([[4]])
print(f"Para X=4, Y={previsto[0]:.2f}")
```

# Exibição de dados e reta

```python

# Plotar os dados e a reta de regressão
plt.scatter(X, Y, color='blue', label='Dados reais')
plt.plot(X, y_pred, color='red', linewidth=2, label='Reta de regressão')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Regressão Linear")
plt.show()
```
